<i> 1.1 </i> Инициализируйте полносвязную сеть нормальным шумом N(0, 0.1) с архитектурой 784 -> 500 x (10 раз) -> 10. В качестве активации возьмите tanh

```python
def init_layer(layer, a, b, mytype):
    # Тут надо быть аккуратным — можно случайно создать копию и менять значения у копии
    weight = layer.state_dict()['weight']
    bias = layer.state_dict()['bias']
    bias.zero_()
    if (mytype=="normal"):
        weight.normal_(mean=a, std=b)
    elif (mytype=="uniform"):
        weight.uniform_(a,b)

def forward_hook(self, input_, output):
    std = input_[0].std().item()
    print('forward', std)

def backward_hook(self, grad_input, grad_output):
    std = grad_input[0].std().item()
    print('backward', std)
    
def make_linear(in_dim, out_dim):
    layer = nn.Linear(in_dim, out_dim)
    layer.register_forward_hook(forward_hook)
    layer.register_backward_hook(backward_hook)
    return layer
    
```

<i>1.2 Пропустите батч изображений через нейронную сеть и вычислите дисперсию активаций. Затем вычислите градиент и получите дисперсию градиентов. Сравните эти значения между собой для разных слоев.</i>

```python
weights_var='dumb'
sizes=[28*28]+[500]*20+[10]
layers=[]
for in_dim, out_dim in zip(sizes,sizes[1:]):
    layer=make_linear(in_dim, out_dim)
    if weights_var=='dumb':
        init_layer(layer, 0,0.1,"normal")
    else:
        raise RuntimeError("improper initialization")
    layers.append(layer)

network = nn.Sequential(*layers)

#пример:
n_objects = 100
X = images_train[:n_objects].view(n_objects, -1).data
y = labels_train[:n_objects].data
activations = network(X)
loss_fn = torch.nn.NLLLoss()
optimizer = torch.optim.Adam(network.parameters(), lr=0.001) 
loss = loss_fn(activations, y)
loss.backward()
```

### Часть 2: Dropout

Другим полезным слоем является Dropout. В нем с вероятностью 1-p зануляется выход каждого нейрона. Этот слой уже реализован в pyTorch, поэтому вновь реализовывать его не интересно. Давайте реализуем DropConnect — аналог Dropout. В нем с вероятностью 1-p зануляется каждый вес слоя.

<i> 2.1 Реализуйте линейный слой с DropConnect </i>

```python
# полезная функция: .bernoulli_(p)
# не забывайте делать requires_grad=False у маски
# помните, что в вычислениях должны участвовать Variable, а не тензоры

class DropConnect(nn.Module):
    def __init__(self, input_dim, output_dim, p=0.5):
        super(DropConnect, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim)
        self.p = p #вероятность занулить нейрон

    def forward(self, x):
        P=torch.ones_like(self.linear.weight)*(1-self.p)
        if self.training:
            mask = P.bernoulli()
        else:
            mask = P

        mask=mask.detach() #то же самое что mask.requires_grad=False
        mask = mask.data
        output = F.linear(x, self.linear.weight * mask, self.linear.bias)
        return output
```

<i> 
2.2 Сравните графики обучения нейроных сетей:
1. Свертки из TestNetwork -> 128 -> 128 -> 10 с ReLU и Dropout между всеми слоями 
2. Свертки из TestNetwork -> 128 -> 128 -> 10 с ReLU DropConnect вместо всех линейных слоев
</i>

```python
class TestNetwork(nn.Module):
    def __init__(self, final_part):
        super().__init__()    
        
        channels = 1
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(channels, 2, 3, padding=1),    
            nn.MaxPool2d(2),
            nn.ReLU(),
            nn.Conv2d(2, 4, 3, padding=1),            
            nn.MaxPool2d(2),
            nn.ReLU(),
        )
        
        #input_size = 7 * 7 * 4 = 196
        self.flatten = nn.Flatten()
        
        self.final_part = final_part
        
        self.log_softmax = nn.LogSoftmax(1)        
        
    def forward(self, x):
        x = self.conv_layers(x)
        x = self.flatten(x)
        x = self.final_part(x)
        return self.log_softmax(x)
```

```python
mode='none'
sizes=[196,128,128,10]
layers=[]
for in_dim, out_dim in zip(sizes, sizes[1:]):
    if mode=='drop_connect':
        layer=DropConnect(in_dim, out_dim,0.2)
        layers.append(layer)
    else:
        layer=nn.Linear(in_dim, out_dim)
        layers.append(layer)
    if mode=='dropout':
        layer=nn.Dropout(0.2)
        layers.append(layer)
    layers.append(nn.ReLU())


from train_utils import train

network = TestNetwork(nn.Sequential(*layers))

train(network, train_loader, test_loader, 10, 0.001, device='cpu')
```

```python
mode='dropout'
sizes=[196,128,128,10]
layers=[]
for in_dim, out_dim in zip(sizes, sizes[1:]):
    if mode=='drop_connect':
        layer=DropConnect(in_dim, out_dim,0.2)
        layers.append(layer)
    else:
        layer=nn.Linear(in_dim, out_dim)
        layers.append(layer)
    if mode=='dropout':
        layer=nn.Dropout(0.2)
        layers.append(layer)
    layers.append(nn.ReLU())
    
network = TestNetwork(nn.Sequential(*layers))

train(network, train_loader, test_loader, 10, 0.001, device='cpu')
```

```python
mode='drop_connect'
sizes=[196,128,128,10]
layers=[]
for in_dim, out_dim in zip(sizes, sizes[1:]):
    if mode=='drop_connect':
        layer=DropConnect(in_dim, out_dim,0.2)
        layers.append(layer)
    else:
        layer=nn.Linear(in_dim, out_dim)
        layers.append(layer)
    if mode=='dropout':
        layer=nn.Dropout(0.2)
        layers.append(layer)
    layers.append(nn.ReLU())
    
network = TestNetwork(nn.Sequential(*layers))

train(network, train_loader, test_loader, 10, 0.001, device='cpu')
```

В test-time стохастичность Dropout убирают и заменяют все веса на их ожидаемое значение: $\mathbb{E}w = pw + (1-p)0 = pw$.

### Часть 3: Batch Normalization

Наконец, давайте рассмотрим Batch Normalization. Этот слой вычитает среднее и делит на стандартное отклонение. Среднее и дисперсия вычисляются по батчу независимо для каждого нейрона. У этого слоя есть две важные проблемы: его нельзя использовать при обучении с размером батча 1 и он делает элементы батча зависимыми. Давайте реализуем аналог батч нормализации: <a href=https://arxiv.org/pdf/1607.06450.pdf>Layer normalization</a>. В layer normalization среднее и дисперсия вычисляются по активациям нейронов, независимо для каждого объекта.

<i> 3.1 Реализуйте Layer Normalization </i>

```python
# полезные функции: .std(dim), .mean(dim)

class LayerNormalization(nn.Module):
    def forward(self, x):
        mu=x.mean(axis=1).reshape(-1,1)
        stdd=((x-mu)**2).mean(axis=1).reshape(-1,1)
        stdd=stdd**0.5
        output = (x-mu)/stdd
        return output
```

<i> 
3.2 Сравните графики обучения нейроных сетей:
1. Свертки из TestNetwork -> 128 -> 128 -> 10 с ReLU и Batch normalization между всеми слоями 
2. Свертки из TestNetwork -> 128 -> 128 -> 10 с ReLU и Layer normalization между всеми слоями 
</i>

```python
##### YOUR CODE HERE #####
mode='batchnorm'
sizes=[196,128,128,10]
layers=[]
for in_dim, out_dim in zip(sizes, sizes[1:]):
    layer=nn.Linear(in_dim, out_dim)
    layers.append(layer)
    if mode=='batchnorm':
        layer=nn.BatchNorm1d(out_dim)
        layers.append(layer)
    elif mode=='layernorm':
        layer=LayerNormalization()
        layers.append(layer)
    layers.append(nn.ReLU())

network = TestNetwork(nn.Sequential(*layers))

train(network, train_loader, test_loader, 10, 0.001, device='cpu')
```

```python
mode='layernorm'
sizes=[196,128,128,10]
layers=[]
for in_dim, out_dim in zip(sizes, sizes[1:]):
    layer=nn.Linear(in_dim, out_dim)
    layers.append(layer)
    if mode=='batchnorm':
        layer=nn.BatchNorm1d(out_dim)
        layers.append(layer)
    elif mode=='layernorm':
        layer=LayerNormalization()
        layers.append(layer)
    layers.append(nn.ReLU())

network = TestNetwork(nn.Sequential(*layers))

train(network, train_loader, test_loader, 10, 0.001, device='cpu')
```
